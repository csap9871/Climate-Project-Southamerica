### Atmospheric circulation - description of the features

The climate in South America is influenced by three major features of the atmospheric circulation (Oliver, 2008) (cf. seasonal geopotential and wind maps):

That is first, the equatorial Intertropical Convergence Zone (ITCZ), which oscillates between a southern (austral summer) and northern (austral winter) position over the course of a year. The seasonal (meridional) shifting of the high geopotential belt of the upper troposphere can give a hint about this ITCZ oscillation. However, its' influence is mainly visible in a seasonal shift of the precipitation pattern (cf. following section: Precipitation).  
At the north-east coast, the movement of the ITCZ is also visible in a shift of wind direction from north-easterly in January to east/south-easterly in July (with quickly decreasing windspeeds inland)(cf. following section graphics: surface wind).

Secondly, two semi-permanent, subtropical anticyclones (south-eastern Pacific and southern Atlantic) influence west and east coast of South America. Like the ITCZ they find their most northern position during austral winter months (and vice vers). While subsiding air of the Pacific High leads to dry conditions at the subtropical west coast, the Atlantic High brings precipitation to the east coast (between northern Argentina and southern Brazil).  
On a global scale it can be seen that the strength of semi-permanent Highs and Lows is generally stronger in the winter months (of each hemisphere) (cf. Exercise 08-General Circulation). This phenomenon is clearly visible on the monthly mean sea level pressure maps shown below: During winter months, especially the Atlantic High shows enhanced magnitude and spatial coverage.  
Both Highs induce anticyclonic surface winds (i.e. anti-clockwise in the Southern Hemisphere) that have an effect on the coastal regions.

Finally, southern South America is dominated by the all-year-round westerly winds, that hit the coast of Chile south of the Pacific High and therefore lead to orographic precipitation. They result from the pressure difference between the subtropical high-pressure belt and the subpolar low-pressure belt.

#### Visualization of atmospheric drivers with a meridional cross-section
The January and July atmospheric cross-sections show zonally averaged ($-90$ - $-30$) values for zonal, meridional and vertical wind components, which support the analysis of the atmospheric circulation given above.  
The meridional advance and retreat of zonal and vertical wind visualize the annual oscillation of the ITCZ and suggest a range from 10S (austral summer) to 10N (austral winter), in accord with Waliser and Gautier (1993) as well as with the precipitation analysis that follows (Hovmller plot). The equatorial trade winds (southeastly-northeastly) can be seen as a combination of the westward flow (blue shading) with the ITCZ-oriented meridional wind vectors. In austral summer, the westward flow at and south of the ITCZ is attenuated due to the semi-permanent Atlantic High (cf. section surface winds above). Instead, the northerly flow (low-level-jet east of the Andes) endures and brings moist air southward (Liebmann et al, 2004). According to Liebmann et al (2004) the low-level jet is a result of the northeasterly trade winds that get redirected southward by the Andes, in combination with "heating on a sloping terrain". In austral winter, subsiding air (Hadley circulation) leads to dry subtropical conditions (cf. section precipitation - Hovmller plot). In the figure above (July), the Hadley circulation seems to be manipulated by the strong westerlies, especially aloft.  
The shift of the prevailing westerlies (as dicussed above) is clearly visible; so is the increase of wind speed with height, as compared to the circulation plots earlier. The Pacific High limits the meridional extent of the westerly flow in the lower troposphere, while aloft maximum values of wind speed can be found.

#### Monthly mean temperature 
Equatorial and tropical South America experience high average temperatures throughout the whole year. South of 20S monthly averages show a gradual decline to negative values in the winter months as latitude in- and insolation decreases. Summers are hot in the subtropics, but the narrower continent south of 50S experiences more cool temperate conditions due to a maritime influence (Oliver, 2008).  
West coast temperatures are tempered by the Peru Current that transports cold water northward as part of the South Pacific gyre. Upwelling of cold ocean water enhances this phenomenon (Zhou et al, 1998). The opposite effect takes place along the east coast, where the Brazil Current ships warm water southward (South Atlantic gyre) (Brink and Robinson, 2005).  
A decrease of mean temperatures along the Andes come hand in hand with an increased elevation.

#### Description of an annual cycle of the subregions
Northwest (equatorial) Brazil lies within the influence of the ITCZ througout the whole year and therefore shows an almost constant temperature cycle. According to Grimm (2010), the maximum rainfall of the wet season occurs in Central Brazil in December/January/February and then travels northward. Therefore northwest Brazil experiences its wettest season March through April, whereas (austral) winter and spring are the relative dry seasons with the ITCZ being at its northernmost position. Note that there is still a high amount of precipitation during that perod.  
The Altiplano lies at the tropical-subtropical border. Grimm (2010) continues in accord with the atmospheric cross-section from above and the precipitation cycle shown here, that winter is the dry season of the tropical belt. Summerly insolation combined with a water vapor advection due to the northerly low-level jet (cf. atmospheric cross-section) lead to a convective summer with maximum amounts of precipitation.  
Its unexciting temperature cycle (warm in summer, cold in winter) indicates the high elevation of the Altiplano: The triple-border region Argentina-Brazil-Paraguay experiences a similar cycle with 10-13C higher temperatures, although it is located several degrees further south.  
In the latter region precipitation falls all-year round, with a slight enhancement in March-April and September-October. Grimm (2010) speculates that this is "due to the seasonal change of the upper-tropospheric subtropical jet, and the interaction with the northerly low-level jet east of the Andes." The surface wind field suggests that some of that rainfall might be due to a water vapor advection by the Atlantic Anticyclone.
